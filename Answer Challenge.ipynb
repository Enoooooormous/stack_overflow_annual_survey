```python
# import the packages needed for the project

import numpy as np
import pandas as pd
import random
```

```python
# importing the cvs data

path = 'D:\\Ame\\base_de_respostas_10k_amostra.csv'
master_data = pd.read_csv(path,decimal=',')
```

```python
# first visualization and sanity check

nbr_records = master_data.shape[0]
nbr_attributes = master_data.shape[1]

assert nbr_records == 10000
assert nbr_attributes == 129

master_data.head()
```

```python
# identifying there are missing values in the columns used in later analysis

category_col = ['Country', 'Hobby', 'OpenSource', 'ConvertedSalary', 'OperatingSystem', 'LanguageWorkedWith', 'CommunicationTools', 'CompanySize'] 
for cat_col in category_col:
    print(cat_col + ":", sum(master_data[cat_col].isna()))
```

```python
# create the operating system table (list of unique operating system with unique ID)

# get the list of unique os
list_of_os = master_data.OperatingSystem.unique().astype('str')

# create the unique ID number for each os
x = random.randint(0, 10000)
ids = range(x, x+len(list_of_os))
ids_array = [i for i in ids]

# create the dataframe
data_tuples = list(zip(list_of_os, ids_array))
os_df = pd.DataFrame(data_tuples, columns=['os', 'os_ID'])

# visualize the dataframe
print(os_df.head())

# perform sanity check
assert os_df.shape[0] == len(master_data.OperatingSystem.unique())
assert len(os_df.os_ID.unique()) == os_df.shape[0]

# set the country_ID column as index
os_df.set_index('os_ID', inplace=True)

os_df
```

```python
# export the table for further analysis in SQL

desktop_path = 'C:\\Desktop\\'
table_name = 'sistema_operacional.csv'
os_df.to_csv(desktop_path+table_name)
```

```python
# create the country table (list of unique country with unique ID)

# get the list of unique countries
list_of_countries = master_data.Country.unique().astype('str')

# create the unique ID number for each country
x = random.randint(0, 10000)
ids = range(x, x+len(list_of_countries))
ids_array = [i for i in ids]

# create the dataframe
data_tuples = list(zip(list_of_countries, ids_array))
country_df = pd.DataFrame(data_tuples, columns=['country', 'country_ID'])

# visualize the dataframe
print(country_df.head())

# perform sanity check
assert country_df.shape[0] == len(master_data.Country.unique())
assert len(country_df.country_ID.unique()) == country_df.shape[0]

# set the country_ID column as index
country_df.set_index('country_ID', inplace=True)

country_df
```

```python
# export the table for further analysis in SQL

table_name = 'pais.csv'
country_df.to_csv(desktop_path+table_name)
```

```python
# create the company size table (list of unique company size with unique ID)

# get the list of unique company size
list_of_company_size = master_data.CompanySize.unique().astype('str')

# create the unique ID number for each company size
x = random.randint(0, 10000)
ids = range(x, x+len(list_of_company_size))
ids_array = [i for i in ids]

# create the dataframe
data_tuples = list(zip(list_of_company_size, ids_array))
company_size_df = pd.DataFrame(data_tuples, columns=['company_size', 'company_size_ID'])

# visualize the dataframe
print(company_size_df.head())

# perform sanity check
assert company_size_df.shape[0] == len(master_data.CompanySize.unique())
assert len(company_size_df.company_size_ID.unique()) == company_size_df.shape[0]

# set the country_ID column as index
company_size_df.set_index('company_size_ID', inplace=True)

company_size_df
```

```python
# export the table for further analysis in SQL

table_name = 'empresa.csv'
company_size_df.to_csv(desktop_path+table_name)
```

```python
### Alternatively, the process can be automatized ###

# use of a loop to create a dictionary of dataframe
columns = ['Country', 'OperatingSystem', 'CompanySize']

# create an empty dictionary to host the dataframes
dict_df = []

# create the loop pulling the individual columns, extract the individual categories, create a unique ID and store the whole in a dataframe
for col in columns:
    # get the list of unique values
    list_of_unique_values = master_data[col].unique().astype('str')

    # create the unique ID number for each value
    x = random.randint(0, 10000)
    ids = range(x, x+len(list_of_unique_values))
    ids_array = [i for i in ids]

    # create the dataframe
    data_tuples = list(zip(list_of_unique_values, ids_array))
    name_df = col + '_df'
    name_cols = [col, col+'_ID']
    d = {col: pd.DataFrame(data_tuples, columns=name_cols)}
    
    # sanity check
    assert d[col].shape[0] == len(master_data[col].unique())
    assert len(d[col][name_cols[1]].unique()) == d[col].shape[0]
    
    # store in an intermediary dictionary
    dict_df.append(d)
    
# extract the dataframes to create individual tables
country_df = dict_df[0]['Country'].set_index('Country_ID')
os_df = dict_df[1]['OperatingSystem'].set_index('OperatingSystem_ID')
company_size_df = dict_df[2]['CompanySize'].set_index('CompanySize_ID')
```

```python
# create the respondent table (list of attributes for each respondent with unique ID)

# start with getting each respondent ID from the master data
respondents = np.array(master_data.Respondent)
respondent_df = pd.DataFrame(respondents, columns=['respondent_id'])
```

```python
# create the column name

# concatenate respondent_ to the id number
name = 'respondent_'
list_of_names = [name + str(id) for id in respondent_df.respondent_id]
respondent_df['name'] = list_of_names

# check that each respondent name is unique
assert len(respondent_df.name.unique()) == len(respondent_df.respondent_id.unique())
```

```python
# create the column contribution to open source project

# convert string categorical to binary
open_source = np.where(master_data.OpenSource == 'Yes', 1, 0)
assert len(open_source) == respondent_df.shape[0]

# add it to the dataframe
respondent_df['contrib_open_source'] = open_source
```

```python
# create the column programming as a hobby

# convert string categorical to binary
hobby = np.where(master_data.Hobby == 'Yes', 1, 0)
assert len(hobby) == respondent_df.shape[0]

# add it to the dataframe
respondent_df['programma_hobby'] = hobby
```

```python
### Alternatively, the process can be automatized ###

# select the columns to binarize
columns = ['OpenSource', 'Hobby']

# create the loop 
for col in columns:
    # take the individual columns and convert 'Yes' in ones and 'No' in zeros
    binary_col = np.where(master_data[col] == 'Yes', 1, 0)
    
    # sanity check
    assert len(binary_col) == respondent_df.shape[0]

    # add it to the dataframe
    respondent_df[col] = binary_col
    
respondent_df
```

```python
# create the column salary

# convert string to float and replace nan values by 0
salaries = pd.to_numeric(master_data.ConvertedSalary).fillna(0)

# convert dollars to reais 
rate = 3.81
monthly_salaries = (salaries * rate) / 12

### This step would require more analysis (is not consistent with the Salary column). 
### The business rule mentioned that the ConvertedSalary was the converted annual salary in US Dollars. 
### Therefore we have to multiply it by the exchange rate (dollar is stronger than real) and divide by 12 to get monthly salary. 

assert len(monthly_salaries) == respondent_df.shape[0]


# add it to the dataframe
respondent_df['salario'] = monthly_salaries

respondent_df
```

```python
# create a intermediary table to respect the integrity of the data

mini_master_data = master_data[['Respondent', 'OperatingSystem', 'Country', 'CompanySize']]
mini_master_data.columns = ['respondent_id', 'operating_system', 'country', 'company_size']
mini_master_data
```

```python
# import the operating system ID (to serve as a foreign key)

# get the number of columns for verification
prev_nbr_col = len(respondent_df.columns)

# add a operating_system column in the dataframe using a join to match with respondent_id 
respondent_df = pd.merge(respondent_df, mini_master_data, how='inner', left_on='respondent_id', right_on='respondent_id')

# merge the respondent_df and os_df to get the os_ID
respondent_df = pd.merge(respondent_df, os_df.reset_index(), how='inner', left_on='operating_system', right_on='os').drop(columns=['operating_system', 'os'])

# check that only one column was added
act_nbr_col = len(respondent_df.columns)
assert act_nbr_col == prev_nbr_col + 1

respondent_df
```

```python
# import the country ID (to serve as a foreign key)

# get the number of columns for verification
prev_nbr_col = len(respondent_df.columns)

# add a country column in the dataframe using a join to match with respondent_id 
respondent_df = pd.merge(respondent_df, mini_master_data, how='inner', left_on='respondent_id', right_on='respondent_id').drop(columns=['operating_system', 'company_size'])

# merge the respondent_df and country_df to get the country_ID
respondent_df = pd.merge(respondent_df, country_df.reset_index(), how='inner', left_on='country', right_on='country').drop(columns=['country'])

# check that only one column was added
act_nbr_col = len(respondent_df.columns)
assert act_nbr_col == prev_nbr_col + 1

respondent_df
```

```python
# import the company_size ID (to serve as a foreign key)

# get the number of columns for verification
prev_nbr_col = len(respondent_df.columns)

# add a company_size column in the dataframe using a join to match with respondent_id 
respondent_df = pd.merge(respondent_df, mini_master_data, how='inner', left_on='respondent_id', right_on='respondent_id').drop(columns=['operating_system', 'country'])

# merge the respondent_df and country_df to get the country_ID
respondent_df = pd.merge(respondent_df, company_size_df.reset_index(), how='inner', left_on='company_size', right_on='company_size').drop(columns=['company_size'])

# check that only one column was added
act_nbr_col = len(respondent_df.columns)
assert act_nbr_col == prev_nbr_col + 1

respondent_df.set_index('respondent_id', inplace=True)

respondent_df
```

```python
# export the table for further analysis in SQL

table_name = 'respondente.csv'
respondent_df.to_csv(desktop_path+table_name)
```

```python
# create a table inventorying all the programming languages used by each respondent

# split the respondent used languages over several columns (one language per column)
respondent_languages = master_data.LanguageWorkedWith.str.split(";", n = 40, expand = True)

# get the respondent id
respondent_languages['respondent_id'] = master_data.Respondent

# reshape the dataframe to have only one language column
respondent_languages = respondent_languages[respondent_languages.columns].set_index('respondent_id').stack().reset_index(name='language')
respondent_languages = respondent_languages.drop(columns='level_1')
respondent_languages.set_index('respondent_id')

respondent_languages
```

```python
# create the programming language table (list of unique programming language with unique ID)

# get the list of unique languages
list_of_languages = respondent_languages.language.unique().astype('str')

# create the unique ID number for each language
x = random.randint(0, 10000)
ids = range(x, x+len(list_of_languages))
ids_array = [i for i in ids]

# create the dataframe
data_tuples = list(zip(list_of_languages, ids_array))
language_df = pd.DataFrame(data_tuples, columns=['language', 'language_ID'])

# visualize the dataframe
print(language_df.head())

# perform sanity check
assert language_df.shape[0] == len(respondent_languages.language.unique())
assert len(language_df.language.unique()) == language_df.shape[0]

# set the country_ID column as index
language_df.set_index('language_ID', inplace=True)

language_df
```

```python
# export the table for further analysis in SQL

table_name = 'linguagem_programacao.csv'
language_df.to_csv(desktop_path+table_name)
```

```python
# import the language_ID (to serve as a foreign key)

# get the number of columns for verification
prev_nbr_col = len(respondent_languages.columns)

# merge the respondent_df and country_df to get the country_ID
respondent_languages = pd.merge(respondent_languages, language_df.reset_index(), how='inner', left_on='language', right_on='language').drop(columns=['language'])

# check that only one column was added
act_nbr_col = len(respondent_languages.columns)
assert act_nbr_col == prev_nbr_col

respondent_languages.set_index('respondent_id', inplace=True)

respondent_languages
```

```python
# export the table for further analysis in SQL

table_name = 'resp_usa_linguagem.csv'
respondent_languages.to_csv(desktop_path+table_name)
```

```python
# export the table for further analysis in SQL

table_name = 'ferramenta_comunic.csv'
communication_df.to_csv(desktop_path+table_name)
```
